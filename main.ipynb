# Bubble Popper Game

### Overview

Functionality:
    Gameplay Pieces
        1)A mouse-rotated shooter sprite
        2)A projectile sprite of four types, oriented from the shooter sprite along the mouse orientation, which hits the bubbles
        3)A descending curtain of bubble sprites, which pop when hit by the right projectile
        4)A timer of gameplay, used for scoreboarding and gameplay pacing
        5)A scoreboard of points, saved at each game end, always displayed
        6)A menu for controlling the game and changing theme
            A)Elements- Change Theme, Pause, Restart, Quit, About
                i)Change Theme: Reloads all graphic styles to new theme
                ii)Pause: Interrupts the timer, Shows unpause window
                iii)Restart: Restarts the game, keeping current theme, discards score
                iv)Quit: Closes the program
                v)About: Shows a lovely about box with fun information
        7)An indicator for the current projectile type
        8)A background for the game screen (dependent on theme)
        9)An endgame pop-up window with a retry/quit option, and motivational text
        10)Motivational text: Several statements of encouragement, love, and support 
        11)Main Menu: Shown at program launch and restart.
            A)Elements- Three theme choices presented, clicking one begins the game
    Three Skins- Three visual options for how the game will look.
        1)Bug Theme- A little bug shoots pebbles at mushrooms
        2)Manta Theme- A cool manta ray shoots lightning at bubbles
        3)Skull Theme- A menacing skull shoots blood at eyeballs
